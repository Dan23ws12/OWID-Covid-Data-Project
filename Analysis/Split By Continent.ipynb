DEFINING THE FUNCTION THAT WILL SPLIT THE COVIDDEATHS AND COVIDVACCINATIONS DATABASES BY CONTINENTS

```python
def split_by_continent(df, target_folder):
    continents = coviddeaths["continent"].unique()
    continents = continents[~pd.isnull(continents)]
    no_entry = coviddeaths[~coviddeaths["continent"].isin(continents)]
    no_entry.to_csv(r"C:\Users\danie\Documents\data projects\Covid Data Project\Covid Deaths Data\null_entry.csv", index=False)
    for continent in continents:
        df_by_continent = df[df["continent"] == continent]
        df_by_continent.to_csv(target_folder + "\\" +continent+ ".csv", index = False)  
```

RUNNING THE FUNCTION ON COVIDDEATHS

```python
target_folder = r"C:\Users\danie\Documents\data projects\Covid Data Project\Covid Deaths Data"
split_by_continent(coviddeaths, target_folder)    
```

RUNNING THE FUNCTION ON COVIDVACCINATIONS

```python
covidvaccinations = pd.read_excel(r"C:\Users\danie\Documents\data projects\Covid Data Project\Covid Vaccinations Data\CovidVaccinations.xlsx")
target_folder = r"C:\Users\danie\Documents\data projects\Covid Data Project\Covid Vaccinations Data"
split_by_continent(coviddeaths, target_folder)    
```
