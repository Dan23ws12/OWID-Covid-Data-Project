# Finding The Number Of Times The Values In All Columns Change Per Location

```python
unique_loc = covid_deaths["location"].unique()
covid_deaths_red = covid_deaths.drop(["iso_code", "population", "continent"], axis = 1)
nunique_by_loc_df = pd.DataFrame(data = {}) #empty dataframe, this  dataframe contains the number of unique column values by country
for loc in unique_loc:
    nunique_by_loc = covid_deaths_red[covid_deaths_red["location"] == loc].nunique()
    nunique_by_loc["location"] = loc
    nunique_by_loc_df = pd.concat([nunique_by_loc_df, pd.DataFrame(nunique_by_loc).transpose()], ignore_index = True)
nunique_by_loc_df
```
