```python
def get_irrelevant_col(df):
    return [x for x in df if df[x].count() < 60000]
```

# Creating a Function That Finds The Number Of Times The Values In All Columns Change Per Location

```python
def get_location_rel(df):
    unique_loc = df["location"].unique()
    df_red = df.drop(["iso_code", "population", "continent"], axis = 1)
    nunique_by_loc_df = pd.DataFrame(data = {}) #empty dataframe, this  dataframe contains the number of unique column values by country
    for loc in unique_loc:
        nunique_by_loc = df_red[df_red["location"] == loc].nunique()
        nunique_by_loc["location"] = loc
        nunique_by_loc_df = pd.concat([nunique_by_loc_df, pd.DataFrame(nunique_by_loc).transpose()], ignore_index = True)
    return nunique_by_loc_df
```

# Finding the columns in the covid_deaths dataframe that have less than 60,000 non-null entries

```python
covid_deaths_drop = get_irrelevant_col(covid_deaths)
covid_deaths_drop
```

```python
#removing irrelevant columns
cov_deaths_red = covid_deaths.drop(covid_deaths_drop, axis = 1)
```

```python
cov_deaths_loc_nuniq_df = get_location_rel(cov_deaths_red)
cov_deaths_loc_nuniq_df
```

```python
# A list of names of columns that have at most one unique value per location
one_per_loc_lst = ["median_age", "aged_65_older", "aged_70_older", "gdp_per_capita", "cardiovasc_death_rate", "diabetes_prevalence", "female_smokers",
                  "hospital_beds_per_thousand", "life_expectancy", "population_density", "human_development_index", "iso_code", "continent"]
```

```python
#table with columns from covid_deaths that has only one value per location
location_df = cov_deaths_red.loc[:, ["location"] + one_per_loc_lst]
location_df.head(6)
```

```python
remaining_col = cov_deaths_red.columns
#list containing names columns from covid_deaths that are about death
death_df_col, other_df_col = [], []
for col in remaining_col:
    if re.search(".*death.*", col):
        death_df_col.append(col)
    else:
        other_df_col.append(col)

death_df = cov_deaths_red.loc[:, ["location", "date"] + death_df_col]
death_df.head(6)
```

```python
other_df = cov_deaths_red.loc[:, other_df_col]
other_df = other_df.drop(columns = ["iso_code", "continent"])
other_df.head(6)
```
